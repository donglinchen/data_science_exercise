Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python https://matplotlib.org

For details about Matplotlib, refer to https://matplotlib.org/stable/Matplotlib.pdf

**Figure** is top level container for all the plot elements, figure contains **Axes**, Axes contains two or more Axis, Axis contains data limits, ticks, ticklabels...

Everything you can see on the figure is an **Artist**, including **Text** objects, **Line2D** objects, **collections** objects...




matplotlib.pyplot is a state-based interface to matplotlib. It provides a MATLAB-like way of plotting.

pyplot is mainly intended for interactive plots and simple cases of programmatic plot generation

Instead of using pyplot.plot to create plot directly, We can also use pyplot to create Figure object, add Axes to figure, then call the methods of axes to plot data, this is the Object Oriented way of plotting object.

Create a Figure object, add axes to figure object and plot a line:

```python
y = np.sin(np.linspace(0, 2 * np.pi))
fig = plt.figure()
ax = fig.add_axes([0,0,0.76,0.76])
ax.plot(y)
```

Use pyplot.subplots() to create both figure and axes objects, and then call the object methods to plot data:

```python
fig, ax = plt.subplots()
ax.plot(y)
```

A another way is using pyplot.plot to plot directly on implicit axes, this is intended for interactive plots and simple cases of programmatic plot generation

```python
plt.plot(y)
```

Set title, xlabel, ylable, legend on **explicit** axes

```python
fig, ax = plt.subplots()
ax.plot(y, label='linear')
# ax.set_title('Title')
# ax.set_xlabel('X Label')
# ax.set_ylabel('Y Label')
ax.set(title='Title', xlabel='X Label', ylabel='Y Label')
ax.legend()
```

Set title, xlabel, ylable, legend on implicit axes using pyplot

```python
plt.plot(y, label='linear')
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.legend()
```

An axes can be attached with more than one artist like Line2D object

```python
x = np.linspace(0,2, 20)
fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.legend()
```

Create 2 x 2 subplots, plot hist, line, scatter, hist2d plots

```python
data = np.random.randn(2, 100)
fig, axs = plt.subplots(2, 2, figsize=(5, 5))
axs[0, 0].hist(data[0])
axs[0, 1].plot(data[0], data[1])
axs[1, 0].scatter(data[0], data[1])
axs[1, 1].hist2d(data[0], data[1])
plt.show()
```

Set range of the x and y axes, with red dashes

```python
# object-oriented (OO) style
fig, ax = plt.subplots()
ax.plot(y, 'r--')
ax.set_xlim(0, 25)
ax.set_ylim(0,1)
```

```python
# pyplot functions
plt.plot(y, 'r--')
plt.axis([0, 25, 0, 1])
```

```python
fig, ax = plt.subplots()
ax.plot(x, x)
ax.annotate("(1,1)", xy=(1,1), xytext=(0.5, 1.5), arrowprops={'facecolor':'black'})
```
